```python
# importing requests , beautiful soup
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=realme+mobile&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&as-pos=1&as-type=RECENT&suggestionId=realme+mobile%7CMobiles&requestId=5845d07c-68ca-4baa-a815-0bd46fe62063&as-backfill=on"
page = requests.get(url)
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
titles = soup.find_all('div',class_="_4rR01T")
ratings = soup.find_all('div',class_="gUuXy-")
reviews = soup.find_all('span',class_="_2_R_DZ")
prices = soup.find_all('div',class_="_30jeq3 _1_WHN1")
```

```python
t = []
r = []
re = []
p = []

for title,rating,review,pri in zip(titles,ratings,reviews,prices):
    t.append(title.text)
    r.append(rating.text)
    re.append(review.text)
    p.append(pri.text)
    
d = {'titles':t,'ratings':r,'reviwes':re,'prices':p}
df = pd.DataFrame(data=d)
df.to_csv("flipkart_realme_data.csv",index=False)
```
