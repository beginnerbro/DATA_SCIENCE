```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("H:\\company_sales_data.csv")
print(df)
```

# Exercise 1: Read Total profit of all months and show it using a line plot

```python
x = df["month_number"]
y = df["total_profit"]
plt.xlabel("month number")
plt.ylabel("total profit")
plt.plot(x,y)
plt.show()
```

# Get total profit of all months and show line plot with the following Style properties

```python
x = df["month_number"]
y = df["total_profit"]
plt.xlabel("month number")
plt.ylabel("total profit")
plt.plot(x,y,"o:r",mfc="black",linewidth="3",label="profit data of last year")
plt.legend(loc="lower right")
plt.show()
```

# Read all product sales data and show it using a multiline plot¶

```python
fc = df["facecream"]
fw = df["facewash"]
tp = df["toothpaste"]
bs = df["bathingsoap"]
sh = df["shampoo"]
mo = df["moisturizer"]
plt.plot(fc,color="blue",marker="o",label="face cream sales data")
plt.plot(fw,color="orange",marker="o",label="face wash sales data")
plt.plot(tp,color="green",marker="o",label="toothpaste sales data")
plt.plot(bs,color="red",marker="o",label="bathing soap sales data")
plt.plot(sh,color="violet",marker="o",label="shampoo sales data")
plt.plot(mo,color="brown",marker="o",label="mositure sales data")
plt.legend(loc="upper left")
plt.title("sales data")
plt.xlabel(" sales units in number ")
plt.ylabel(" month number ")
```

# Read toothpaste sales data of each month and show it using a scatter plot

```python
tp = df["toothpaste"]
month = df["month_number"]
plt.grid(True,linewidth="3",linestyle="--")
plt.scatter(month,tp,label="tooth paste sales data")
plt.xlabel("month number")
plt.ylabel(" number of units sold")
plt.title(" tooth paste sales data ")
plt.legend(loc="upper left")
```

# Read face cream and facewash product sales data and show it using the bar chart

```python
fc = df["facecream"]
fw = df["facewash"]
month = df["month_number"]
plt.bar(month,fc,label="face cream sales data")
plt.bar(month,fw,label="face wash sales data")
```
